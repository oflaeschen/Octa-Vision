```python
import cv2
import pandas as pd
import numpy as np
from ultralytics import YOLO
from tracker import *



model=YOLO('yolov8s.pt')
```

```python
def RGB(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE :  
        colorsBGR = [x, y]
        print(colorsBGR)
        

cv2.namedWindow('RGB')
cv2.setMouseCallback('RGB', RGB)
```

```python
cap=cv2.VideoCapture('vidyolov8.mp4')

my_file = open("coco.txt", "r")
data = my_file.read()
class_list = data.split("\n")
#print(class_list)
count=0

tracker = Tracker()
```

```python
#desenhando o ROI
area = [(270,238),(294, 280),(592,226),(552,207)]
#PARA CONF. em cenários podemos alterar a área ou melhorar a detecção
```

```python
area_c = set()
while True:
    
    ret,frame = cap.read()
    if not ret:
        break
    count += 1
    if count % 3 != 0:
        continue
    frame=cv2.resize(frame,(1020,500))

    results=model.predict(frame)

    #    print(results)
    a=results[0].boxes.boxes
    px=pd.DataFrame(a).astype("float")
#    print(px)
    list=[]
             
    
    for index,row in px.iterrows():
        #print(row)
        
        x1=int(row[0])
        y1=int(row[1])
        x2=int(row[2])
        y2=int(row[3])
        d=int(row[5])
        c=class_list[d]

        #Detectando apenas carros.
        #se tivermos uma classe == Carros...
        if 'car' in c:
            list.append([x1,y1,x2,y2]) #criando a lista com as info desse carro. 
    bbox_id = tracker.update(list)
    for bbox in bbox_id: #vamos desenhar os retangulos nos objetos
        x3,y3,x4,y4,id = bbox #coordenadas do tracker
        #achando o ponto central
        cx = int(x3+x4)//2
        cy = int(y3+y4)//2
        results=cv2.pointPolygonTest(np.array(area,np.int32), ((cx,cy)),False) #nosso ponto está dentro do poligono?
        #results = +1 ou -1  

        if results >= 0: #se realmente existir um ponto no retangulo, fazemos algumas funcoes
            #só desenharemos se o carro estiver dentro da ROI
            cv2.circle(frame,(cx,cy),4,(0,0,255),-1)
            cv2.rectangle(frame,(x3,y3),(x4,y4),(0,0,255),2)
            cv2.putText(frame,str(id),(x3,y3), cv2.FONT_HERSHEY_COMPLEX,0.5,(255,0,0),1)
            area_c.add(id)  ## add as ids numa lista
        
    cv2.polylines(frame, [np.array(area,np.int32)],True,(255,255,0),3) #desenhando o poligono na tela
    count=len(area_c)
    cv2.putText(frame,str(count),(61,146), cv2.FONT_HERSHEY_PLAIN,5,(255,255,255),1)
    cv2.imshow("RGB", frame)
    if cv2.waitKey(1)&0xFF==27:
        break

cap.release()
cv2.destroyAllWindows()
```
